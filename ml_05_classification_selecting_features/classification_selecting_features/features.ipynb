```python
import pandas as pd
```

```python
data = pd.read_csv('data/hotel.csv')
data
```

```python
data.isna().sum()
```

```python
data[data.duplicated()]
```

```python
data = data.drop_duplicates()
data.reset_index(drop=True, inplace=True)
```

```python
data.duplicated().sum()
```

---
Treinando sem explorar as Features

```python
x = data.drop(columns='booking_status')
x
```

```python
y = data[['booking_status']]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(random_state=42)
```

```python
rfc.fit(x_train, y_train)
```

```python
rfc.score(x_test, y_test)
```

Feature importance

```python
importancia = pd.DataFrame({'colunas': x_train.columns, 'importancia': rfc.feature_importances_})
best_importancias = importancia.sort_values(by='importancia', ascending=False).head(20)
best_importancias
```

---
SelectFromModel

```python
from sklearn.feature_selection import SelectFromModel
```

```python
sfm = SelectFromModel(rfc, max_features=10, prefit=True)
```

```python
x_train_selecionado = sfm.transform(x_train)
x_test_selecionado = sfm.transform(x_test)
```

```python
rfc_selecionado = RandomForestClassifier(random_state=42)
```

```python
rfc_selecionado.fit(x_train_selecionado, y_train)
```

```python
rfc_selecionado.score(x_test_selecionado, y_test)
```

---

```python
from sklearn.model_selection import GridSearchCV
```

```python
parametros_busca = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}
```

```python
grid_search = GridSearchCV(estimator=rfc,
                           param_grid=parametros_busca,
                           scoring='accuracy',
                           cv=5)
```

```python
grid_search.fit(x_train, y_train)
```

```python
grid_search.cv_results_
```

```python
grid_search.best_estimator_
```

```python
rfc_grid_selecionadas = SelectFromModel(grid_search.best_estimator_, max_features=10, prefit=True)
```

```python
x_train_grid_selecionado = rfc_grid_selecionadas.transform(x_train)
y_train_grid_selecionado = rfc_grid_selecionadas.transform(x_test)
```

```python
rfc_grid_model = grid_search.best_estimator_
```

```python
rfc_grid_model.fit(x_train_grid_selecionado, y_train)
```

```python
rfc_grid_model.score(x_test_selecionado, y_test)
```

---

```python
from sklearn.metrics import confusion_matrix, roc_curve, auc
```

```python
y_pred = rfc_grid_model.predict(x_test_selecionado)
```

```python
matriz_confusao = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(matriz_confusao, annot=True, cmap='coolwarm', fmt='d');
```

```python
fpr, tpr, treshoulds = roc_curve(y_test, y_pred)
```

```python
plt.plot(fpr, tpr, color='darkorange', lw=2)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
```

```python
roc_auc = auc(fpr, tpr)
roc_auc
```

```python
parametros = {'max_depth': None,
 'min_samples_leaf': 1,
 'min_samples_split': 5,
 'n_estimators': 300,
 'random_state': 42}
```

```python
rfc_rfe = RandomForestClassifier(**parametros)
```

```python
selecionador = RFE(rfc_rfe, n_features_to_select=10, step=1)
```

```python
selecionador.fit(x_train, y_train)
```

```python
x_train_selecionado_rfe = selecionador.transform(x_train)
x_test_selecionado_rfe = selecionador.transform(x_test)
```

```python
rfc_rfe.fit(x_train_selecionado_rfe, y_train)
```

```python
rfc_rfe.score(x_train_selecionado_rfe, y_train)
```

```python
y_pred = rfc_rfe.predict(x_test_selecionado_rfe)
```

```python
matriz_confusao_rfe = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(matriz_confusao_rfe, annot=True, cmap='coolwarm', fmt='d')
```

---

```python
rfc_rfecv = RandomForestClassifier(**parametros)
```

```python
from sklearn.feature_selection import RFECV
from sklearn.model_selection import StratifiedKFold
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
rfecv = RFECV(
    estimator=rfc_rfecv,
    step=1,
    cv=cv,
    scoring='roc_auc',
    min_features_to_select=1,
    importance_getter='auto'
)
```

```python
rfecv.fit(x_train, y_train)
```

```python
rfecv.n_features_
```

```python
rfecv.ranking_
```

```python
rfecv.cv_results_
```

```python
y_pred = rfecv.predict(x_test)
```

```python
matriz_confusao_rfecv = confusion_matrix(y_test, y_pred)
sns.heatmap(matriz_confusao, annot=True, cmap='coolwarm', fmt='d')
```

```python
resultado = rfecv.cv_results_
```

```python
resultado['mean_test_score'].mean()
```

```python
resultado_df = pd.DataFrame({'Valores': resultado['mean_test_score']})
```
