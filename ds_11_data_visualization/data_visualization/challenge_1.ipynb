```python
import pandas as pd
```

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
df = pd.DataFrame(index=lojas, data=vendas_2022)
df
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(2,2, figsize=(12,7))
fig.subplots_adjust(hspace=0.5, wspace=0.2)
fig.suptitle("Store sales trend", x=0.0, ha='left', fontsize=20)

axs[0,0].plot(df.loc['A'], lw=2, color='brown')
axs[0,0].set_title("Empresa A", fontsize=16)

axs[0,1].plot(df.loc['B'], lw=2, color='slateblue')
axs[0,1].set_title("Empresa B", fontsize=16)

axs[1,0].plot(df.loc['C'], lw=2, color='orange')
axs[1,0].set_title("Empresa C", fontsize=16)

axs[1,1].plot(df.loc['D'], lw=2, color='forestgreen')
axs[1,1].set_title("Empresa D", fontsize=16)

for ax in axs.flat:
    ax.set_xlabel("Month")
    ax.set_ylabel("Sales")
    ax.set_ylim(0, 500)

plt.show()
```

```python
colors = ['darkviolet', 'green', 'darkblue', 'coral']
```

```python
fig, axs = plt.subplots(2,2, figsize=(12,7))
fig.subplots_adjust(hspace=0.5, wspace=0.2)
fig.suptitle("Store sales trend", x=0.0, ha='left', fontsize=20)

for i, ax in enumerate(axs.flat):
    ax.plot(df.loc[df.index[i]], color=colors[i], lw=3)
    ax.set_title(f'Vendas na loja {df.index[i]}', loc='left', fontsize=16)
    ax.set_xlabel('Ms', fontsize=14)
    ax.set_ylabel('Nmero de vendas', fontsize=14)
    ax.tick_params(labelsize=12)
    ax.grid(color='lightgrey')
    
for ax in axs.flat:
    ax.set_xlabel("Month")
    ax.set_ylabel("Sales")
    ax.set_ylim(0, 500)
    
plt.show()
```
